```python
import pandas as pd
```

```python
df_institutes = pd.read_csv("project/College_Scorecard_Raw_Data_05192025/Most-Recent-Cohorts-Institution.csv")
```

```python
df_institutes
```

```python
best_10 = [
    'UNITID', 'INSTNM', 'CITY', 'STABBR',
    'TUITFTE', 'DEBT_MDN', 'PCTPELL',
    'MD_EARN_WNE_P10', 'MN_EARN_WNE_P10',
    'MEDIAN_HH_INC'
]

df_institutes_best = df_institutes[best_10]
df_institutes_best.head()
```

```python
df_programs = pd.read_csv("project/College_Scorecard_Raw_Data_05192025/Most-Recent-Cohorts-Field-of-Study.csv")
```

```python

print(df_programs.columns.tolist())
```

```python
best_cols = [
    'UNITID', 'INSTNM', 'CONTROL', 'CIPCODE', 'CIPDESC', 'CREDLEV', 'CREDDESC',
    'DEBT_ALL_STGP_ANY_MDN', 'DEBT_ALL_PP_ANY_MDN',
    'EARN_MDN_1YR', 'EARN_MDN_4YR', 'EARN_MDN_5YR',
    'EARN_PELL_WNE_MDN_1YR', 'EARN_NOPELL_WNE_MDN_1YR',
    'BBRR2_FED_COMP_DLNQ',
    'DISTANCE'
]
df_programs_best = df_programs[best_cols]
```

```python
df_programs_best
```

```python
df_combined = df_institutes_best.merge(
    df_programs_best,
    on="UNITID",
    how="inner"    # use inner or left depending on your goal
)
```

```python
df_combined
```

```python
df_clean = df_combined.dropna()
```

```python
print("Before:", df_combined.shape)
print("After:", df_clean.shape)
```

```python
df_clean.isna().sum()
df_clean.isna().mean().sort_values(ascending=False)
```

```python
df_clean.describe().T
```

```python
df_clean.select_dtypes(include='object').describe().T
```

```python
threshold = 0.20   # 20% PS values allowed (change to 0.30 if needed)
ps_pct = (df_clean == "PS").mean()

cols_to_drop_ps = ps_pct[ps_pct > threshold].index
df_clean = df_clean.drop(columns=cols_to_drop_ps)
print("After dropping PS columns:", df_clean.shape)
```

```python
df_clean.select_dtypes(include='object').describe().T
```

```python
df_clean.nunique().sort_values(ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cols = [df_clean.columns[i] for i in range(df_clean.shape[1]) if df_clean.dtypes[i] != 'object']
for c in cols:
    sns.histplot(df_clean[c], kde=True)
    plt.title(f"Distribution of {c}")
    plt.show()
```
